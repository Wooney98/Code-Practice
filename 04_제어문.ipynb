# 조건문
- 특정 조건을 만족(`True`)하면 코드를 실행
- `if`, `elif`, `else`
    - `if` : 조건이 참이면 코드를 실행해라
    - `elif` : else if, 그게 아니고 `elif`의 조건이 참이면 실행해라
    - `else` : `if`와 `elif` 모두 조건이 참이 아니면 실행해라

- 조건문 구조
```python
if<condition>:
    code context
elif <condition>:
    code context
...
...
...
else:
    code context

```

````python
acc = 0.92
if acc >= 0.9:
    print("최고입니다.")
elif acc < 0.9 and acc >=0.8:
    print("보통입니다.")
elif acc < 0.8 and acc >= 0.7:
    print("별로입니다.")
else:
    print("나쁩니다.")
````

- 중앙값
    - 주어진 값을 크기순으로 순서대로 정렬했을때 가장 중앙에 위치한 값
    - 이상치에 영향을 받지 않는다.

````python
model_scores1 = [0.92 , 0.83 , 0.91 , 0.77, 0.99 , 0.61 , 0.88,0.76,0.75] #  홀수 개수
model_scores2 = [0.92 , 0.83 , 0.91 , 0.77, 0.99 , 0.61 , 0.88,0.76,0.75 , 0.51] # 짝수 개수
print(len(model_scores1))
print(len(model_scores2))
````

- 조건문을 이용하여 중앙값을 추출하는 코드 

````python
#짝수일 경우
scores = model_scores1
scores_sort = sorted(scores)
idx = 0
median = 0

if len(scores_sort) %2==0:
    idx=len(scores_sort)//2
    median=(scores_sort[idx-1]+ scores_sort[idx])/2
else:
    idx = len(scores_sort)//2+1
    median = scores_sort[idx]


print("중간값은 {}입니다.".format(median))
````

## `while`
- 조건이 참(`True`)인 동안 반복
- 무한 반복 주의!!!

```python
while <condition>:
    code context

```

- 3단 출력해보기

````python
n = 3
i = 0
while i <10:
    print(f"{n} X {i} = {n*i}")
    i += 1
````

- `break` : 해당 반복문을 빠져 나온다.

````python
n = 9
i = 1
while(True):
    print(f"{n} X {i} = {n*i}")
    i += 1

    if i > 9:
        break
````

- `continue` : 진행을 중단하고 다음 반복으로 넘어간다. 

````python
n = 3
i = 1
while i < 10:
    if i % 2 != 0:
        i+=1
        continue

    print(f"{n} X {i} = {n*i}")
    i+=1
````

```
age_list = [20,42,None,30,25,None,27,60,50]
None 값들은 결측치
None 값을 제외한 나이값들의 평균을 구하시오.
```

````python
age_list = [20,42,None,30,25,None,27,60,50]
index = len(age_list)
    
tmp = []

while index:
    index -= 1
    if age_list[index] is not None:
        tmp.append(age_list[index])
avg = sum(tmp)/len(tmp)
avg = int(avg)
avg
````

```

```

````python
index = len(age_list)
while index:
    index -= 1
    if age_list[index] is None:
        age_list[index] = avg
age_list
````

-중첩 while문을 이용하여 구구단 출력하기
```
2 단입니다.
2 X 1 = 2
...
...
9 단입니다.
9 X 1 = 9
...
...
...
9 X 9 = 81
```

````python
n = 2

while n <10:  # 2단부터 시작해서 2~9범위 지정
    
    print(f"{n} 단입니다.") 
    i = 1  # i 를 1로 시작하게 하고
    
    while i < 10:  # 1~9 범위 지정
        print(f"    {n} X {i} = {n*i}") 
        i += 1  # 출력한 후에 i에 1을 반복해서 더해 9까지 출력한 뒤 
    
    n += 1  # n에 1을 더해서 다시 3번줄 부터 반복한다.
````

- while 문으로 1~100까지 짝수만 더한 값 출력하기

````python
i = 0
result = 0

while i < 101:
    i=i+1 # i의 값이 1씩 증가 하면서,
    if i%2 == 0: # 나머지가 0일때,
        result = result + i # result에 i를 반복해서 더한다.
print(f"{result}")
````

- break문 활용

````python
j=0
result2 = 0

while True:
    if j > 100:
        break
    
    j = j+1
    if j%2==0:
        result2 = result2+j

print(f"{result2}")
````

- continue문 활용

````python
k = 0
result3 = 0

while k < 101:
    k = k+1

    if k%2 != 0:
        continue
    result3 = result3 + k

print(f"{result3}")
````

# `for` 
- 반복 가능(iterable)한 객체를 더이상 꺼낼수 없을때까지 반복.
- `list`, `tuple`, `set`, `dict`, `str`

```python
for <variable> in <iterable>:
    code context

```

````python
lst = [1,2,3,4]

for item in lst:
    print(item**2)
````

- for문을 이용한 9단

````python
n = 9
lst = [1,2,3,4,5,6,7,8,9]
print("9단")

for i in lst:
    result = n*i
    print(f"{n} X {i} = {result}")
````
